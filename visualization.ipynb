### Cross Validation

```python
def plot_cross_validation(fold_labels):
  # set height
  plt.figure(figsize=(6,4))

  # set y-axis limits
  plt.ylim([0, 1.5])

  plt.xticks(fold_labels)

  # plot
  plt.plot(fold_labels, accuracy, color = 'limegreen')
  plt.plot(fold_labels, loss, color = 'turquoise')
  plt.plot(fold_labels, recall, color = 'deepskyblue')
  plt.plot(fold_labels, precision, color = 'royalblue')

  # add plot title
  plt.title("Cross Validation Results")
  plt.xlabel("Fold")
  plt.ylabel("Score")
  plt.legend(["Accuracy", "Loss", "Recall", "Precision"], ncol = 4, loc = 'upper center')

  plt.show()
```
